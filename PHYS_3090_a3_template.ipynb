```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fourier_series(x, nmax):
    n = np.arange(1, nmax+1)
    return(2.*np.sum(np.power(-1.,n+1)*np.sin(n*x)/n))

def taylor_series(x, nmax):
    n = np.arange(1, nmax+1,2)
    return(np.pi/2 - (4./np.pi) * np.sum(np.cos(n*x)/np.power(n,2)))
```

```python
x_vals = np.linspace(-np.pi, np.pi, 1000)
for nmax in [5,10,50]:
    plt.plot(x_vals, [fourier_series(x, nmax) for x in x_vals],label="$N = $"+str(nmax), lw=2)
plt.plot(x_vals, x_vals, '--', lw=2)
plt.legend()
plt.show()
```
