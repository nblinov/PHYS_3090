```python
import numpy as np
from scipy.special import kn

import matplotlib.pyplot as plt

```

```python
x = np.logspace(-2,5,100)



plt.plot(np.sqrt(np.pi/(2.*x)),'--',label=r'$K_n(x)e^{x}$ via asymptotics')
plt.plot(np.exp(x)*kn(0,x),label=r'$K_n(x)e^x$ via scipy')

plt.xscale('log')
plt.yscale('log')
plt.legend()
```
