```python
import numpy as np
import math
import scipy
import matplotlib.pyplot as plt
```

```python
def sho_wavefunction(x,n):
  return(np.exp(-0.5*x**2)*scipy.special.eval_hermite(n, x))
```

```python
x_list = np.linspace(-5,5,100)

for n in range(0,3):
  plt.plot(x_list, sho_wavefunction(x_list, n),label='$n=$'+str(n))
#plt.plot(x_list,np.power(x_list,2),'--')
plt.ylim([-3,3])
plt.title("Quantum Harmonic Oscillator Wavefunctions")
plt.xlabel(r"$x$")
plt.ylabel(r"$\psi_n(x)$")
plt.legend()
plt.show()
```

```python
def hermite_delta_representation(x, nmax):
    n = np.arange(0, nmax+1)
    normalization = np.sqrt(np.pi)*np.power(2.,n)*scipy.special.factorial(n)
    return(np.exp(-x**2)*np.sum(scipy.special.eval_hermite(n, x)*scipy.special.eval_hermite(n, 0)/normalization))
```

```python
x_vals = np.linspace(-5,5, 1000)

for nmax in [5,10,100]:
    plt.plot(x_vals, [hermite_delta_representation(x, nmax) for x in x_vals],label="$N = $"+str(nmax))

plt.legend()

plt.title('Hermite Polynomial Completeness Check')
plt.xlabel(r'$x$')
plt.ylabel(r'$\delta_N(x)$')

plt.show()
```

```python
def legendre_delta_representation(x, nmax):
    n = np.arange(0, nmax+1)
    normalization = 2./(2.*n+1)
    return(np.sum(scipy.special.eval_legendre(n, x)*scipy.special.eval_legendre(n, 0)/normalization))
```

```python
x_vals = np.linspace(-1,1, 1000)

for nmax in [5,10,50]:
    plt.plot(x_vals, [legendre_delta_representation(x, nmax) for x in x_vals],label="$N = $"+str(nmax))

plt.legend()

plt.title('Legendre Polynomial Completeness Check')
plt.xlabel(r'$x$')
plt.ylabel(r'$\delta_N(x)$')

plt.show()
```
