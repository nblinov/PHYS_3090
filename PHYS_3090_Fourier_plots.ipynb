```python
import matplotlib.pyplot as plt
import numpy as np

```

```python

def sawtooth_fourier(x, nmax):
    n = np.arange(1, nmax+1)
    return(np.pi/2. - np.sum(np.sin(2.*n*x)/n))

def odd_sawtooth_fourier(x, nmax):
    n = np.arange(1, nmax+1)
    return(2.*np.sum(np.power(-1.,n+1)*np.sin(n*x)/n))

def even_sawtooth_fourier(x, nmax):
    n = np.arange(1, nmax+1,2)
    return(np.pi/2 - (4./np.pi) * np.sum(np.cos(n*x)/np.power(n,2)))
```

```python
x_vals = np.linspace(-np.pi, np.pi, 1000)
def f(x):
  return x % (np.pi)
for nmax in [5,10,50]:
    plt.plot(x_vals, [sawtooth_fourier(x, nmax) for x in x_vals],label="$N = $"+str(nmax), lw=2)
plt.plot(x_vals, f(x_vals), 'k--', lw=2)
plt.legend()

plt.title('Sawtooth Fourier Series')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')

plt.show()
```

```python
x_vals = np.linspace(-np.pi, np.pi, 1000)
def f(x):
  return x
for nmax in [5,10,50]:
    plt.plot(x_vals, [odd_sawtooth_fourier(x, nmax) for x in x_vals],label="$N = $"+str(nmax),lw=2)
plt.plot(x_vals, f(x_vals), 'k--', lw=2)
plt.title('Odd Sawtooth Fourier Series')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend()
plt.show()
```

```python
x_vals = np.linspace(-np.pi, np.pi, 1000)
def f(x):
  return np.abs(x)
plt.plot(x_vals, f(x_vals), 'k--', lw=2)
for nmax in [5,10,50]:
    plt.plot(x_vals, [even_sawtooth_fourier(x, nmax) for x in x_vals],label="$N = $"+str(nmax))

plt.title('Even Sawtooth Fourier Series')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend()
plt.show()
```
