```python
import numpy as np
import matplotlib.pyplot as plt

# Evenly spaced points on the unit circle
n_points = 50
angles = np.linspace(-np.pi, np.pi, n_points, endpoint=True)
circle_points = np.exp(1j * angles)
sqrt_circle = np.sqrt(circle_points)

Re_vals = np.real(sqrt_circle)
Im_vals = np.imag(sqrt_circle)


# Plot the unit circle in complex plane
plt.figure()
plt.plot(np.real(circle_points), np.imag(circle_points), 'o')
plt.gca().set_aspect("equal", adjustable="box")
plt.xlabel(r"$\Re z$")
plt.ylabel(r"$\Im z$")
plt.title("Evenly spaced points on the unit circle")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# Combined plot with labels
plt.figure()
plt.plot(angles, Re_vals, label=r"$\Re \sqrt{z}$")
plt.plot(angles, Im_vals, label=r"$\Im \sqrt{z}$")
plt.xlabel(r"$\arg z$ (radians)")
plt.title(r"Re and Im parts of $\sqrt{z}$ on the unit circle vs argument")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the square root function on a grid
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y
F = np.sqrt(Z)

ReF = np.real(F)
ImF = np.imag(F)


# Real part
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection="3d")
ax1.plot_surface(X, Y, ReF, linewidth=0, antialiased=True)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("Re sqrt(z)")
ax1.set_title("Real part of sqrt(z)")
plt.show()

# Imag part
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection="3d")
ax2.plot_surface(X, Y, ImF, linewidth=0, antialiased=True)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("Im sqrt(z)")
ax2.set_title("Imaginary part of sqrt(z)")
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def sqrt_with_rotated_cut(z: np.ndarray, theta0: float) -> np.ndarray:
  return np.sqrt(np.absolute(z))*np.exp(1j*(np.angle(z) + np.pi + theta0)/2)

theta0 = np.deg2rad(45)
x = np.linspace(-2, 2, 300)
y = np.linspace(-2, 2, 300)
X, Y = np.meshgrid(x, y, indexing="xy")
Z = X + 1j * Y

F = sqrt_with_rotated_cut(Z, theta0)
ReF = np.real(F)
ImF = np.imag(F)

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection="3d")
ax1.plot_surface(X, Y, ReF, linewidth=0, antialiased=True)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("Re sqrt z")
ax1.set_title(f"Real part of rotated-cut sqrt (theta0 = {np.rad2deg(theta0):.1f})")
plt.show()

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection="3d")
ax2.plot_surface(X, Y, ImF, linewidth=0, antialiased=True)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("Im sqrt z")
ax2.set_title(f"Imag part of rotated-cut sqrt (theta0 = {np.rad2deg(theta0):.1f})")
plt.show()
```
